```python
import pandas as pd
import re
```

```python
df_ml_ratings = pd.read_csv('../Sandbox/MovieLens/ml-latest-small/ratings.csv')
df_ml_ratings
```

```python
# Remove the timestamp column
df_ml_ratings = df_ml_ratings.drop(columns=['timestamp'])
df_ml_ratings
```

```python
# Count the number of userIds
num_users = df_ml_ratings.userId.nunique()
print(num_users)
```

```python
# Load the data from the file project_survey_data.csv
df = pd.read_csv('../Sandbox/project_survey_data.csv')
df
```

```python
# Add sequential user IDs starting from a specific number as first column
start_id = num_users + 1  # Starting user ID; note that num_users is zero-based
df['userId'] = range(start_id, start_id + len(df))
df
```

```python
# Organize df by adding a new row for each movie from column headings rated by each user
# Ignore the Timestamp column
df = df.drop(columns='Timestamp')
df_expanded = df.melt(id_vars='userId', var_name='movieId', value_name='rating')
df_expanded['movieId'] = df_expanded['movieId'].astype('string')
df_expanded
```

```python
# Remove rows where 'rating' column has NaN values
df_expanded_clean = df_expanded.dropna(subset=['rating'])

df_expanded_clean
```

```python
# Remove just the square brackets but keep their contents
df_expanded_clean['movieId'] = df_expanded_clean['movieId'].replace(r'\[([^\]]*)\]', r'\1', regex=True)
print(df_expanded_clean.dtypes)
```

```python
# Load the movie dataset
df_ml_movies = pd.read_csv('../Sandbox/MovieLens/ml-latest-small/movies.csv')
df_ml_movies['title'] = df_ml_movies['title'].astype('string')
df_ml_movies
df_ml_movies.dtypes
```

```python
# Create a mapping dictionary from title to movieId in df_ml_movies
title_to_movieid = dict(zip(df_ml_movies['title'], df_ml_movies['movieId']))
print(title_to_movieid)

df_expanded_clean['original_column'] = df_expanded_clean['movieId']

# Replace movieId values in df_expanded based on title
df_expanded_clean['movieId'] = df_expanded_clean['original_column'].map(title_to_movieid)

# find closest match for movieId based on title using get_close_matches
# from difflib library
from difflib import get_close_matches

def get_closest_match(title):
    matches = get_close_matches(title, title_to_movieid.keys())
    if len(matches) > 0:
        return title_to_movieid[matches[0]]
    else:
        return None
    
df_expanded_clean['movieId'] = df_expanded_clean['original_column'].apply(get_closest_match)

df_expanded_clean
```

```python
# Remove original_column column
df_expanded_clean = df_expanded_clean.drop(columns='original_column')
df_expanded_clean
```

```python
# Combine df_ml_ratings and df_expanded_clean to make the complete dataset
df_combined = pd.concat([df_ml_ratings, df_expanded_clean])
df_combined
```

```python
# Save to pickle (binary format)
df_combined.to_pickle('combined_survey_data.pkl')

# Read back pickle file
# df = pd.read_pickle('combined_survey_data.pkl')
```
